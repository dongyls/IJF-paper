### This file contains the confirmation of the content in `/Reproduction/best_hps.csv`.

In the csv file, the four columns are seed number, LSTM units, Dense units, and learning rate, respectively.

```python
import os
import tensorflow as tf

# This file requires TF = 2.9.0a
print(tf.config.list_physical_devices('GPU'), tf.__version__)

# Get the current directory.
current_dir = os.getcwd()

# Set up random seeds for data splitting.
split_rs = [290, 150, 266, 78, 148, 133, 155, 135, 178, 241]
```

### Learning rates:

```python
for rs in split_rs:

    trained_lstm_model = tf.keras.models.load_model(current_dir+'/TrainedModels/ANN_model_' + str(rs) +'.h5')
    print('==========Seed', rs, '==========')
    print('Learning rates:', trained_lstm_model.optimizer.get_config()['learning_rate'])
```

### LSTM and Dense units:

Note that the LSTM unit is the second output shape of layer `lstm_1 (LSTM) `. The Dense unit is the second output shape of layer `dense_2 (Dense)`.

```python
for rs in split_rs:

    trained_lstm_model = tf.keras.models.load_model(current_dir+'/TrainedModels/ANN_model_' + str(rs) +'.h5')
    print('==========Seed', rs, '==========')
    trained_lstm_model.summary()
```
